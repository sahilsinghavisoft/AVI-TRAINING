```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('insurance_data.csv')
```

```python

x = df.iloc[:,:-1].values
y = df.iloc[:,1].values
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=42)
```

```python
model=LogisticRegression()
```

```python
model.fit(x, y)
```

```python

y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
model.score(x_test,y_test)
```

```python
plt.scatter(x_test,y_test,color= 'g', marker='D')
plt.scatter(x_test,y_pred,color='blue', marker='o')
plt.xlabel("Age of person")
plt.ylabel("Bought Insurance 1=Bought 0=Did not Buy")
```

```python
from mlxtend.evaluate import bias_variance_decomp
result = bias_variance_decomp(
    estimator=model,
    X_train=x_train,
    y_train=y_train,
    X_test=x_test,
    y_test=y_test,
    random_seed=42,
    loss='mse'  
)
print("Bias-Variance Decomposition Result:", result)
```

2.Use sklearn.datasets iris flower dataset to train your model using logistic regression. You need to figure out accuracy of your model and use that to predict different samples in your test dataset. In iris dataset there are 150 samples containing following features,Sepal LengthSepal WidthPetal LengthPetal Width
Using above 4 features you will clasify a flower in one of the three categories,SetosaVersicolourVirginica

```python
from sklearn.datasets import load_iris
```

```python
x,y=load_iris(return_X_y=True)
```

```python
x
```

```python
y
```

```python
target_names = ['Setosa', 'Versicolor', 'Virginica']
import pandas as pd
df = pd.DataFrame(data=x, columns=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
df['Target'] = [target_names[i] for i in y]
sns.set(style="whitegrid")
for feature in df.columns[:-1]:
    plt.figure(figsize=(8, 6))
    sns.stripplot(x='Target', y=feature, data=df, jitter=True, palette='Set1')
    plt.title(f'{feature} vs. Target')
    plt.xlabel('Target')
    plt.ylabel(feature)
    plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.discriminant_analysis import StandardScaler


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

```

```python

logistic_model = LogisticRegression()
logistic_model.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = logistic_model.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from mlxtend.evaluate import bias_variance_decomp
result = bias_variance_decomp(
    estimator=model,
    X_train=X_train_scaled,
    y_train=y_train,
    X_test=X_test_scaled,
    y_test=y_test,
    random_seed=42,
    loss='mse'  
)
print("Bias-Variance Decomposition Result:", result)
```

```python
new_samples = np.array([[5.1, 3.5, 1.4, 0.2],  
                        [6.2, 2.9, 4.3, 1.3],  
                        [7.7, 3.8, 6.7, 2.2]]) 
new_samples_scaled = scaler.transform(new_samples)
new_predictions = logistic_model.predict(new_samples_scaled)
print("Predictions for new samples:", new_predictions)
```
