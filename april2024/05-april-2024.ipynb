```python
from sklearn.datasets import load_iris
```

 this inbuilt dataset is already clean so i have done here direct sepration of feature and target value with one liner inbuilt function

```python
x,y=load_iris(return_X_y=True)
```

visualizing the features with target values

```python
from matplotlib import pyplot as plt


plt.figure(figsize=(8, 6))
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='viridis', s=50, alpha=0.7)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width')
plt.colorbar(label='Species')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

Using DecisionTreeClassifier model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DecisionTreeClassifier=DecisionTreeClassifier(random_state=42)
```

```python
DecisionTreeClassifier.fit(x_train,y_train)
```

```python
prediction=DecisionTreeClassifier.predict(x_test)
```

```python
print(prediction)
```

```python
print(y_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(prediction,y_test)
```

```python
accuracy
```

Using random forest model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RandomForestClassifier=RandomForestClassifier()
```

```python
RandomForestClassifier.fit(x_train,y_train)
```

```python
prediction1=RandomForestClassifier.predict(x_test)
```

```python
print(prediction1)
```

```python
accuracy1=accuracy_score(prediction1,y_test)
```

```python
accuracy1
```

this mlxtend is one library use to calculate mse and bias and variance .we can also find bias varaince manually but there is a problem of shuffling.

```python
from mlxtend.evaluate import bias_variance_decomp
```

```python
mse,bias,variance=bias_variance_decomp(
    estimator=RandomForestClassifier,
    X_train=x_train,
    X_test=x_test,
    y_train=y_train,
    y_test=y_test,
    random_seed=42
)
print(mse)
print(bias)
print(variance)
```
