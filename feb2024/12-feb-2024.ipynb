```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df=pd.read_csv("gas_prices.csv")
plt.xticks(df.Year[::3])
plt.plot(df['Year'],df['USA'],'b.-',label="United states")
plt.plot(df['Year'],df['Canada'],'r.-')
plt.plot(df['Year'],df['South Korea'],'g.-')
plt.plot(df['Year'],df['Australia'],'y.-')
plt.xlabel("Year")
plt.ylabel("gas value")
plt.legend()
  


```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
x2=np.arange(0,4.5,0.5)
plt.plot(x,y,'b^--',label="2x")
plt.xticks([0,1,2,3,4])
plt.plot(x2[:6],x2[:6]**2,'r',label="2^x")
plt.plot(x2[5:],x2[5:]**2,"r--")
plt.title("Our first Graph!")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
```
