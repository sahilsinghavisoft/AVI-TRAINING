1. You are given an array 'arr' consisting of 'n' integers which denote the position of a stall.

You are also given an integer 'k' which denotes the number of aggressive cows.

You are given the task of assigning stalls to 'k' cows such that the minimum distance between any two of them is the maximum possible.

Print the maximum possible minimum distance.

Example:
Input: 'n' = 3, 'k' = 2 and 'arr' = {1, 2, 3}

Output: 2

Explanation: The maximum possible minimum distance will be 2 when 2 cows are placed at positions {1, 3}. Here distance between cows is 2.

Sample Input 1 :
6 4
0 3 4 7 10 9


Sample Output 1 :
3


Explanation to Sample Input 1 :
The maximum possible minimum distance between any two cows will be 3 when 4 cows are placed at positions {0, 3, 7, 10}. Here distance between cows are 3, 4 and 3 respectively.


Sample Input 2 :
5 2
4 2 1 3 6


Sample Output 2 :
5


Expected time complexity:
Can you solve this in O(n * log(n)) time complexity?


Constraints :
2 <= 'n' <= 10 ^ 5
2 <= 'k' <= n
0 <= 'arr[i]' <= 10 ^ 9
Time Limit: 1 sec.

```python
def possible(stalls, mid, k):
   #this will check if it's possible to place k cows in the stalls with a minimum distance mid between them.
    cow = 1
    last = stalls[0]
    for i in range(len(stalls)):
        if stalls[i] - last >= mid:
            cow += 1
            last = stalls[i]
    return cow >= k
def cow(stalls, k):
# now it will perform binary search to find the maximum minimum distance between cows.
    stalls = sorted(stalls)
    low = 1
    

    high = max(stalls)
    while low <= high:
        mid = (low + high) // 2
        if possible(stalls, mid, k):
            low = mid + 1
        else:
            high = mid - 1
    return high

stalls = list(map(int, input("Enter the stalls positions separated by space: ").split()))
k = int(input("Enter the number of cows: "))
print(cow(stalls, k))
```

```python
def test(num_days, num_chapters, study_time):
    total_time = sum(study_time)
    max_study_time = max(study_time)
    #We will initialize the lower limit of binary search with max_study_time and the upper limit of binary search with total_time.
    start = max_study_time
    end = total_time
    min_time_required = total_time
    
    while start <= end:
        mid = (start + end) // 2
        
        days = 1
        current_time = 0
        
        for i in range(num_chapters):
            current_time += study_time[i]
            if current_time > mid:
                days += 1
                current_time = study_time[i]
# if the days required to study all the chapters is greater than num_days,then answer is not possible and we want more time to study in a given day so we will compress the range to the right hand side. Else we can study all the chapters and we will update the min_time_required with mid.
    
        if days <= num_days:
            min_time_required = mid
            end = mid - 1
        else:
            start = mid + 1
    
    return min_time_required

# Take input from the user
num_days = int(input("Enter the number of days available: "))
num_chapters = int(input("Enter the number of chapters: "))
study_time = list(map(int, input("Enter the time required for each chapter: ").split()))

# Call the function and print the result
print(test(num_days, num_chapters, study_time))
```

3.You have a binary tree of 'N' unique nodes and a Start node from where the tree will start to burn. Given that the Start node will always exist in the tree, your task is to print the time (in minutes) that it will take to burn the whole tree.



It is given that it takes 1 minute for the fire to travel from the burning node to its adjacent node and burn down the adjacent node.



For Example :

After one minute, Nodes (1, 4, 5) that are adjacent to 3 will burn completely.

```python
from queue import Queue
# Binary tree node class
class BinaryTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
# Function to create parent pointers
def create_parent_pointers(parent, root, start):
    target = None
    q = [root]  # Queue for level order traversal
    while q:  # Continue until the queue is not empty
        f = q.pop(0)  # Dequeue the front node
        if f.data == start:  # Check if the current node is the target node
            target = f
        if f.left:  # Check if the left child exists
            q.append(f.left)  # Enqueue the left child
            parent[f.left] = f  # Set parent pointer for left child
        if f.right:  # Check if the right child exists
            q.append(f.right)  # Enqueue the right child
            parent[f.right] = f  # Set parent pointer for right child
    return target  # Return the target node
# Function to calculate time to burn the tree
def time_to_burn_tree(root, start):
    parent = {}  # Dictionary to store parent pointers
    target = create_parent_pointers(parent, root, start)  # Get the target node
    q = [target]  # Queue for level order traversal starting from the target node
    burn = {target: True}  # Dictionary to mark burned nodes
    cnt = 0  # Counter to keep track of time
    while q:  # Continue until the queue is not empty
        s = len(q)  # Get the size of the current level
        fl = 0  # Flag to indicate if any node is burned in the current level
        for i in range(s):  # Iterate over the nodes in the current level
            temp = q.pop(0)  # Dequeue the front node
            if temp.left and not burn.get(temp.left, False):  # Check if left child exists and not burned
                q.append(temp.left)  # Enqueue the left child
                burn[temp.left] = True  # Mark left child as burned
                fl = 1  # Set flag to 1 indicating burning occurred

            if temp.right and not burn.get(temp.right, False):  # Check if right child exists and not burned
                q.append(temp.right)  # Enqueue the right child
                burn[temp.right] = True  # Mark right child as burned
                fl = 1  # Set flag to 1 indicating burning occurred

            if parent.get(temp) and not burn.get(parent[temp], False):  # Check if parent exists and not burned
                q.append(parent[temp])  # Enqueue the parent node
                burn[parent[temp]] = True  # Mark parent node as burned
                fl = 1  # Set flag to 1 indicating burning occurred

        if fl:  # If any node is burned in the current level
            cnt += 1  # Increment the counter

    return cnt  # return the time required to burn the tree
def take_input():
    root_data = int(input("Enter root node data (-1 for null): "))  # Input root node data
    if root_data == -1:  # Check if root node is null
        start = int(input("Enter start node data: "))  # Input start node data
        return None, start  # Return None for root and start node data

    root = BinaryTreeNode(root_data)  # Create root node
    q = Queue()  # Create a queue for level order traversal
    q.put(root)  # Enqueue the root node
    index = 1  # Initialize index for input list

    while not q.empty():  # Continue until the queue is not empty
        current_node = q.get()  # Dequeue the current node
        
        left_child = int(input(f"Enter left child of {current_node.data} (-1 for null): "))  # Input left child
        if left_child != -1:  # Check if left child exists
            left_node = BinaryTreeNode(left_child)  # Create left child node
            current_node.left = left_node  # Set left child of current node
            q.put(left_node)  # Enqueue the left child
        
        index += 1  # Increment index for input list
        right_child = int(input(f"Enter right child of {current_node.data} (-1 for null): "))  # Input right child
        if right_child != -1:  # Check if right child exists
            right_node = BinaryTreeNode(right_child)  # create right child node
            current_node.right = right_node  # set right child of current node
            q.put(right_node)  # enqueue the right child

        index += 1  # increment index for input list

    start = int(input("Enter start node data: "))  #input start node data
    return root, start  # return root node of the binary tree and start node data
root, start = take_input()  # take input for the binary tree and start node data
print(time_to_burn_tree(root, start))  
```
