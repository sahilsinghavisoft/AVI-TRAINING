```python
import csv
areas = [1000, 1500, 2300, 3540, 4120, 4560, 5490, 3460, 4750, 2300, 9000, 8600, 7100]
with open('areas.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['area'])  
    for area in areas:
        writer.writerow([area])
print("CSV file generated successfully: areas.csv")
```
