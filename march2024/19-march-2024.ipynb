1.Sample problem of predicting home price in monroe, new jersey (USA)
Below is the table containing home prices in monroe twp, NJ. Here price depends on area (square feet), bed rooms and age of the home (in years). Given these prices we have to predict prices of new homes based on area, bed rooms and age.

Find price of home with 3000 sqr ft area, 3 bedrooms, 40 year old

A. 498408.25158031

Find price of home with 2500 sqr ft area, 4 bedrooms, 5 year old

A.578876.03748933

```python
import pandas as pd; 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('homeprices1.csv')
```

```python
df
```

```python
x=df.bedrooms.median()
print(x)
```

```python
#now i willl fill NaN value with median value
df=df.fillna(x)
```

```python
df
```

```python
#split data in two parts input data and output data
x=df.drop('price',axis=1)
print(x)
```

```python
y=df.price
```

```python
y
```

```python
plt.plot(df.area,df.price)
plt.xlabel('area')
plt.ylabel('price')
```

```python
Regression=LinearRegression()
Regression.fit(x,y)
```

```python
Regression.coef_
```

```python
Regression.intercept_
```

```python
prediction=Regression.predict([[3000,3.0,40]])
```

```python
print(prediction)
```

```python
prediction=Regression.predict([[2500,4.0,5]])
```

```python
print(prediction)
```

2.there is hiring.csv. This file contains hiring statics for a firm such as experience of candidate, his written test score and personal interview score. Based on these 3 factors, HR will decide the salary. Given this data, you need to build a machine learning model for HR department that can help them decide salaries for future candidates. Using this predict salaries for following candidates,

2 yr experience, 9 test score, 6 interview score

12 yr experience, 10 test score, 10 interview score

Answer
53713.86 and 93747.79

```python
data=pd.read_csv("hiring.csv")
```

```python
data
```

```python
import math
mean = math.floor(data['test_score(out of 10)'].mean())
print(mean)
```

```python
data['test_score(out of 10)'] = data['test_score(out of 10)'].fillna(mean)
data
```

```python
data.experience=data.experience.fillna('zero')
data
```

```python
data
```

```python
data=data.drop('expereience',axis=1)
```

```python
data
```

```python
replace_dict = {'zero': 0, 'five': 5, 'two': 2, 'seven': 7, 'three': 3, 'ten': 10, 'eleven': 11}
data['experience'] = data['experience'].replace(replace_dict)
```

```python
data
```

```python
x=data.drop('salary($)',axis=1)
```

```python
x
```

```python
plt.plot(x,data['salary($)'])
```

```python
plt.bar(data.experience,data['salary($)'])
plt.xlabel('experience')
plt.ylabel('salary')
```

```python
Regression=LinearRegression()
```

```python
Regression.fit(x,data['salary($)'])
```

```python
prediction1=Regression.predict([[2,9,6]])
```

```python
print(prediction1)
```

```python
prediction2=Regression.predict([[12,10,10]])
```

```python
print(prediction2)
```
